```python
from numpy import ceil
class Dragon():

    def __init__(self, height: int, danger: int, color: str):
        self.height, self.danger, self.color = height, danger, color


    def __str__(self):  # +
        return f'Dragon with height {self.height}, \
danger {self.danger} and color {self.color}'


    def __repr__(self):
        """ возвращaет строку: Dragon(‹рост>, <огнеопасность>, <цвет>) """
        return f'Dragon with height {self.height}, danger {self.danger} \
and color {self.color}.'


    def __add__(self, other):
        new_height = int((self.height + other.height) / 2)
        # рост - среднее арифметическое из двух округлённое до целого вниз
        new_danger = self.danger if self.danger > other.danger else other.danger
        # огнеопасность - большее из двух
        new_color = self.color if self.color[0] < other.color[0] \
                    else other.color
        # цвет меньший по алфавиту
        return Dragon(
            height = new_height, 
            danger = new_danger, 
            color = new_color)


    def __eq__(self, other):  # +
        """ для равенства == """
        if self.height == other.height and\
           self.danger == other.danger and\
           self.color[0] == other.color[0]:
            return True  #  f'{self} И {other} РАВНЫ'
        else:
            return False  # f'{self} И {other} НЕ РАВНЫ'


    def __ne__(self, other):  # +
        """ для неравенства != """
        if self.height != other.height or\
           self.danger != other.danger or\
           self.color[0] != other.color[0]:
            return True  # f'{self} И {other} НЕ РАВНЫ'
        else:
            return False  # f'{self} И {other} РАВНЫ'


    def __lt__(self, other):  # +
        """ для оператора меньше < """
        if self.height < other.height or \
           (self.height == other.height and self.danger < other.danger) or\
           (
               self.height == other.height and self.danger == other.danger
               and self.color[0] < other.color[0]
           ):
            return True  # f'{self} МЕНЬШЕ {other}'
        else:
            return False  # f'{self} НЕ МЕНЬШЕ {other}'
        

    def __gt__(self, other):  # +
        """ для оператора больше > """
        if self.height > other.height or \
          (self.height == other.height and self.danger > other.danger) or\
          (
              self.height == other.height and self.danger == other.danger
              and self.color[0] > other.color[0]
          ):
            return True  # f'{self} БОЛЬШЕ {other}'
        else:
            return False  # f'{self} НЕ БОЛЬШЕ {other}'


    def __sub__(self, number: int):
#       из роста вычитается целая честь от деления роста на число,
# к огнеопасности прибавляется остаток от деления огнеопасности на число
        return Dragon(self.height - (self.height // number), 
                      self.danger + (self.danger % number), 
                      self.color)

  
    def change_color(self, new_color): # +
        self.color = new_color


    def __call__(self, string):
        print(string * self.danger)



dr1 = Dragon(25, 30, 'black')
dr2 = Dragon(25, 30, 'gray')

print(dr1.__dict__)
print(dr1)

print(dr1 < dr2)

dr3 = Dragon(15, 30, 'black')
dr4 = Dragon(25, 30, 'gray')
dr5 = Dragon(75, 47, 'green')

print(dr2, dr3, dr4,  sep="\n")


print(f'{dr3 < dr4 = }')
print(f'{dr3 > dr4 = }')
print(f'{dr4 == dr5 = }')
print(f'{dr3 != dr4 = }')

dr3.change_color('white')
print(f'{dr3 = }')

print(dr1, dr5, sep='\n')

dr7 = dr1 + dr5
print(f'{dr7 = }')

dr7 -= 10

print(f'{dr7 = }')

dr7("Spam")
```
