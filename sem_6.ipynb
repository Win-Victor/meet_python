Даны два массива чисел. Требуется вывести те элементы
первого массива (в том порядке, в каком они идут в первом
массиве), которых нет во втором массиве. 

Пользователь вводит
число N - количество элементов в первом массиве, затем N
чисел - элементы массива. Затем число M - количество
элементов во втором массиве. Затем элементы второго массива


Ввод:


7 


3 1 3 4 2 4 12


6


4 15 43 1 15 1

(каждое число вводится с новой строки)


Вывод:

3 3 2 12



```python
lists = [[], []]
for k in range(len(lists)):
  lists[k] = [int(input(f'Введите {i+1}-е число: ')) for i in range(int(input(f'Сколько цифр в {k+1} массиве: ')))]

list_new = [i for i in lists[0] if i not in lists[1]]
print(list_new)
```

Задача №41. Решение в группах

Дан массив, состоящий из целых чисел. Напишите
программу, которая в данном массиве определит
количество элементов, у которых два соседних и, при
этом, оба соседних элемента меньше данного. Сначала
вводится число N — количество элементов в массиве
Далее записаны N чисел — элементы массива. Массив
состоит из целых чисел.


Ввод: 

5

1 2 3 4 5

(каждое число вводится с новой строки)

Вывод:

0

Ввод:

5

1 5 1 5 1

Вывод:

2


```python
list_ = [int(input(f'Введите {i+1}-е число: ')) for i in range(int(input(f'Сколько цифр в массиве: ')))]

count = 0
for i in range(1, len(list_) - 1):
  if list_[i - 1] < list_[i] > list_[i + 1]:
    count += 1

print(count)
```

Задача 30: Заполните массив элементами арифметической
прогрессии. Её первый элемент, разность и количество
элементов нужно ввести с клавиатуры. Формула для
получения n-го члена прогрессии:

a
n = a1  + (n-1) * d.

Каждое число вводится с новой строки.

Ввод: 7 2 5


Вывод: 7 9 11 13 15



```python
a_1 = int(input('Введите первый элемент: '))
diff_ = int(input('Введите первый разность: '))
count = int(input('Введите сколько нужно элементов: '))

list_ = []
list_.append(a_1)
for i in range(1, count):
  list_.append(a_1 + i * diff_)

print(*list_)
```
